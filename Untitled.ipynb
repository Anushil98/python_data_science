```python
df.sort_values("Very interested",ascending=False,axis=0,inplace=True)
arr=np.array(df)
arr=(arr/2233*100)
df1=pd.DataFrame(arr)
df1=round(df1,2)

df1.index=df.index
df1.columns=df.columns
df1
```

```python
mpl.style.use('default')
ax=df1.plot(kind='bar',figsize=(20,8),width=0.8,color=('#5cb85c','#5bc0de','#d9534f'))
ax.set_title("Percentage of Respondents' Interest in Data Science Areas",fontsize=16)
ax.legend(fontsize=14)
j=0
ax.set_xticklabels(labels=df.index,fontsize=14)
ax.tick_params(top='off', bottom='off', left='off', right='off', labelleft='on', labelbottom='on')
plt.setp(ax.get_yticklabels(), visible=False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

for i in arr:
    x=j-0.35
    ax.text(x,i[0]+1,str(round(i[0],2)),fontsize=14)
    x=j-0.10
    ax.text(x,i[1]+1,str(round(i[1],2)),fontsize=14)
    x=j+0.20
    ax.text(x,i[2]+1,str(round(i[2],2)),fontsize=14)
    j+=1
    
    
```
